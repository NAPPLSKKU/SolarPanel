```python
from scipy.optimize import curve_fit

def piecewise_linear(x, P_dir, P_dif):
    # x0 is the breakpoint
    return np.piecewise(x, [x < 0, x >= 0],
                        [lambda x_val: P_dif,
                          lambda x_val: P_dir * x_val + P_dif])


plt.scatter(np.cos(np.deg2rad(GC1Gamma)), GC1Power, color = 'tab:orange', label = 'Bonneville')
plt.scatter(np.cos(np.deg2rad(cGC1Gamma)), cGC1Power, color = 'tab:blue', label = 'CarPark')
plt.xlabel('cos($\gamma$)')
plt.ylabel('Power Output (W)')
plt.title('Great Circle 1')
plt.ylim(-1,16)
plt.legend()
plt.errorbar(np.cos(np.deg2rad(GC1Gamma)), GC1Power, yerr = GC1stdev, color = 'black', linestyle = 'none')
plt.errorbar(np.cos(np.deg2rad(cGC1Gamma)), cGC1Power, yerr = cGC1stdev, color = 'black', linestyle = 'none')

popt, pcov = curve_fit(piecewise_linear, np.cos(np.deg2rad(cGC1Gamma)), cGC1Power)
P_dir=popt[0]
P_dif=popt[1]
y_fit=piecewise_linear(np.cos(np.deg2rad(cGC1Gamma)),P_dir,P_dif)
plt.plot(np.cos(np.deg2rad(cGC1Gamma)), y_fit,label= "Fitted Curve")
plt.text(-0.95,14,r"$\mathrm{P}_{\mathrm{dir}}$ (W): " +f"{P_dir:.3f}")
plt.text(-0.95,13,r"$\mathrm{P}_{\mathrm{dif}}$ (W): " +f"{P_dif:.3f}")
plt.grid(True)
plt.show()

plt.scatter(np.cos(np.deg2rad(GC2Gamma)), GC2Power, color = 'tab:orange', label = 'Bonneville')
plt.scatter(np.cos(np.deg2rad(cGC2Gamma)), cGC2Power, color = 'tab:blue', label = 'CarPark')
plt.xlabel('cos($\gamma$)')
plt.ylabel('Power Output (W)')
plt.title('Greate Circle 2')
plt.ylim(-1,16)
plt.legend()
plt.errorbar(np.cos(np.deg2rad(GC2Gamma)), GC2Power, yerr = GC2stdev, color = 'black', linestyle = 'none')
plt.errorbar(np.cos(np.deg2rad(cGC2Gamma)), cGC2Power, yerr = cGC2stdev, color = 'black', linestyle = 'none')

popt, pcov = curve_fit(piecewise_linear, np.cos(np.deg2rad(cGC2Gamma)), cGC2Power)
P_dir=popt[0]
P_dif=popt[1]
y_fit=piecewise_linear(np.cos(np.deg2rad(cGC2Gamma)),P_dir,P_dif)
plt.plot(np.cos(np.deg2rad(cGC2Gamma)), y_fit,label= "Fitted Curve")
plt.text(-0.95,14,r"$\mathrm{P}_{\mathrm{dir}}$ (W): " +f"{P_dir:.3f}")
plt.text(-0.95,13,r"$\mathrm{P}_{\mathrm{dif}}$ (W): " +f"{P_dif:.3f}")
plt.grid(True)
plt.show()

plt.scatter(np.cos(np.deg2rad(GC3Gamma)), GC3Power, color = 'tab:orange', label = 'Bonneville')
plt.scatter(np.cos(np.deg2rad(cGC3Gamma)), cGC3Power, color = 'tab:blue', label = 'CarPark')
plt.xlabel('cos($\gamma$)')
plt.ylabel('Power Output (W)')
plt.title('Greate Circle 3')
plt.ylim(-1,16)
plt.legend()
plt.errorbar(np.cos(np.deg2rad(GC3Gamma)), GC3Power, yerr = GC3stdev, color = 'black', linestyle = 'none')
plt.errorbar(np.cos(np.deg2rad(cGC3Gamma)), cGC3Power, yerr = cGC3stdev, color = 'black', linestyle = 'none')

popt, pcov = curve_fit(piecewise_linear, np.cos(np.deg2rad(cGC3Gamma)), cGC3Power)
P_dir=popt[0]
P_dif=popt[1]
y_fit=piecewise_linear(np.cos(np.deg2rad(cGC3Gamma)),P_dir,P_dif)
plt.plot(np.cos(np.deg2rad(cGC3Gamma)), y_fit,label= "Fitted Curve")
plt.text(-0.95,14,r"$\mathrm{P}_{\mathrm{dir}}$ (W): " +f"{P_dir:.3f}")
plt.text(-0.95,13,r"$\mathrm{P}_{\mathrm{dif}}$ (W): " +f"{P_dif:.3f}")
plt.grid(True)
plt.show()

plt.scatter(np.cos(np.deg2rad(GC4Gamma)), GC4Power, color = 'tab:orange', label = 'Bonneville')
plt.scatter(np.cos(np.deg2rad(cGC4Gamma)), cGC4Power, color = 'tab:blue', label = 'CarPark')
plt.xlabel('cos($\gamma$)')
plt.ylabel('Power Output (W)')
plt.title('Greate Circle 4')
plt.ylim(-1,16)
plt.legend()
plt.errorbar(np.cos(np.deg2rad(GC4Gamma)), GC4Power, yerr = GC4stdev, color = 'black', linestyle = 'none')
plt.errorbar(np.cos(np.deg2rad(cGC4Gamma)), cGC4Power, yerr = cGC4stdev, color = 'black', linestyle = 'none')

popt, pcov = curve_fit(piecewise_linear, np.cos(np.deg2rad(cGC4Gamma)), cGC4Power)
P_dir=popt[0]
P_dif=popt[1]
y_fit=piecewise_linear(np.cos(np.deg2rad(cGC4Gamma)),P_dir,P_dif)
plt.plot(np.cos(np.deg2rad(cGC4Gamma)), y_fit,label= "Fitted Curve")
plt.text(-0.95,14,r"$\mathrm{P}_{\mathrm{dir}}$ (W): " +f"{P_dir:.3f}")
plt.text(-0.95,13,r"$\mathrm{P}_{\mathrm{dif}}$ (W): " +f"{P_dif:.3f}")
plt.grid(True)
plt.show()

plt.scatter(np.cos(np.deg2rad(GC5Gamma)), GC5Power, color = 'tab:orange', label = 'Bonneville')
plt.scatter(np.cos(np.deg2rad(cGC5Gamma)), cGC5Power, color = 'tab:blue', label = 'CarPark')
plt.xlabel('cos($\gamma$)')
plt.ylabel('Power Output (W)')
plt.title('Greate Circle 5')
plt.ylim(-1,16)
plt.legend()
plt.errorbar(np.cos(np.deg2rad(GC5Gamma)), GC5Power, yerr = GC5stdev, color = 'black', linestyle = 'none')
plt.errorbar(np.cos(np.deg2rad(cGC5Gamma)), cGC5Power, yerr = cGC5stdev, color = 'black', linestyle = 'none')

popt, pcov = curve_fit(piecewise_linear, np.cos(np.deg2rad(cGC5Gamma)), cGC5Power)
P_dir=popt[0]
P_dif=popt[1]
y_fit=piecewise_linear(np.cos(np.deg2rad(cGC5Gamma)),P_dir,P_dif)
plt.plot(np.cos(np.deg2rad(cGC5Gamma)), y_fit,label= "Fitted Curve")
plt.text(-0.95,14,r"$\mathrm{P}_{\mathrm{dir}}$ (W): " +f"{P_dir:.3f}")
plt.text(-0.95,13,r"$\mathrm{P}_{\mathrm{dif}}$ (W): " +f"{P_dif:.3f}")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axes for 2x3 subplots
fig, axs = plt.subplots(2, 3, figsize=(18, 10), sharey=True)

# Titles and data containers for looping
titles = [
    "GC1: 0° below sun",
    "GC2: 6° below sun",
    "GC3: 12° below sun",
    "GC4: 18° below sun",
    "GC5: 24° below sun"
]

GC_Altitudes = [GC1Altitude, GC2Altitude, GC3Altitude, GC4Altitude, GC5Altitude]
GC_STDEVS = [
    GC1STDEVSolarPanelPower,
    GC2STDEVSolarPanelPower,
    GC3STDEVSolarPanelPower,
    GC4STDEVSolarPanelPower,
    GC5STDEVSolarPanelPower
]
GC_Gammas = [GC1Gamma, GC2Gamma, GC3Gamma, GC4Gamma, GC5Gamma]
GC_Powers = [GC1Power, GC2Power, GC3Power, GC4Power, GC5Power]
cGC_Altitudes = [cGC1Altitude, cGC2Altitude, cGC3Altitude, cGC4Altitude, cGC5Altitude]
cGC_Gammas = [cGC1Gamma, cGC2Gamma, cGC3Gamma, cGC4Gamma, cGC5Gamma]
cGC_Powers = [cGC1Power, cGC2Power, cGC3Power, cGC4Power, cGC5Power]

# Function reused from your code
def piecewise_linear(x, P_dir, P_dif):
    return np.piecewise(x, [x < 0, x >= 0],
                        [lambda x_val: P_dif,
                         lambda x_val: P_dir * x_val + P_dif])

# Plot all GC plots
for i in range(5):
    row, col = divmod(i, 3)
    ax = axs[row, col]

    # Fit curve to CarPark
    popt, pcov = curve_fit(piecewise_linear, np.cos(np.deg2rad(cGC_Gammas[i])), cGC_Powers[i])
    P_dir, P_dif = popt

    # SubPower calculation
    SubPower = []
    for gamma, power in zip(GC_Gammas[i], GC_Powers[i]):
        fit_val = piecewise_linear(np.cos(np.deg2rad(gamma)), P_dir, P_dif)
        SubPower.append(power - fit_val)

    x_vals = np.sin(np.deg2rad(GC_Altitudes[i]))
    y_vals = SubPower

    # Polynomial fit
    x_poly = np.sin(np.deg2rad(cGC_Altitudes[i]))
    y_poly = SubPower
    coeffs = np.polyfit(x_poly, y_poly, 2)
    y_poly_fit = np.polyval(coeffs, x_poly)

    sorted_idx = np.argsort(x_poly)
    x_poly_sorted = x_poly[sorted_idx]
    y_poly_fit_sorted = y_poly_fit[sorted_idx]

    # Plotting
    ax.scatter(x_vals, y_vals, color='tab:green', label="Bonneville - Fit Residual")
    ax.errorbar(
        x_vals,
        y_vals,
        yerr=GC_STDEVS[i] + np.abs(np.sqrt(pcov[0][0]) * x_vals) + np.abs(np.sqrt(pcov[1][1])),
        color='black',
        linestyle='none'
    )
    ax.plot(x_poly_sorted, y_poly_fit_sorted, color='tab:red', linestyle='--',
            label=f'Fit: {np.round(coeffs[0],2)}x² + {np.round(coeffs[1],2)}x + {np.round(coeffs[2],2)}')

    ax.set_title(titles[i], fontsize = 16*1.5)
    ax.set_xlabel("sin(Altitude)", fontsize = 16*1.2)
    if col == 0:
        ax.set_ylabel("Power Output (W)", fontsize = 16*1.2)
    ax.set_ylim(-1, 16)
    ax.set_xlim(-1,1)
    ax.tick_params(axis='both', labelsize=16)
    ax.grid(True)
    ax.legend(fontsize=18)

# Hide unused subplot (6th one)
axs[1, 2].axis('off')

# Adjust layout
plt.tight_layout()
plt.suptitle("Upwelling Bonneville – All Great Circles", fontsize=16, y=1.02)
plt.subplots_adjust(top=0.92)
plt.show()
```
